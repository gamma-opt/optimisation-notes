```julia
using Plots, LaTeXStrings, LinearAlgebra
pyplot()

x = range(0,10, length=2) 
c1(x) = 2 + x  #x2 <= x1 + x2
c2(x) = 1 - x

x1 = (0,2)
x2 = (0,1)
x3 = (1,0)
y = (2,2)
ref = (2,1)

plot(Shape([x1, x2, x3, (5,0), (5,c1(5))]), fill = 0.2, alpha = 0.0,
    xaxis = (L"x_1", (-0.06,3)),
    yaxis = (L"x_2", (-0.06,3)),
    aspect_ratio = :equal,
    bordercolor= "white",
    legend = true,
    label = "",
    xticks = 0:3,
    yticks = 0:3,
    size = (250, 250))

hline!([0], color = :black, label = "")
vline!([0], color = :black, label = "")

plot!(x, c1.(x), color = :1, label = "")
plot!(x, c2.(x),color = :1, label = "")

quiver!([x1[1],x3[1],ref[1],ref[1]],[x1[2],x3[2],ref[2],ref[2]], 
    quiver=([1/2,1/2,1/2,1/2],[1/2,0,1/2,0],), color = :2, label = "recession cone")

scatter!(x1, color = :2, label = "")
scatter!(x2, color = :2, label = "")
scatter!(x3, color = :2, label = "")
scatter!(y, color = :2, label = "")

annotate!(x1[1] + 0.1, x1[2] - 0.1, (L"x_1", 9, :left))
annotate!(x2[1] + 0.1, x2[2] + 0.1, (L"x_2", 9, :left))
annotate!(x3[1], x3[2] + 0.1, (L"x_3", 9, :bottom))
annotate!(x1[1] + 0.55, x1[2] + 0.4, (L"w_1", 9, :left))
annotate!(x3[1] + 0.5, x3[2] + 0.05, (L"w_2", 9, :bottom))
annotate!(y[1] + 0.1, y[2], (L"y", 9, :left))
annotate!(ref[1] + 0.25, ref[2] - 0.25, ("Recession \ncone", 7, :above))

savefig("Figure5.pdf")
```
